```python
import csv
 
dates_year = []
dates_month = []
dates_day = []
airT = []
 
with open('BB_AirT_10_17.csv') as csvDataFile:
    csvReader = csv.reader(csvDataFile)
    
    for row in csvReader:
        dates_year.append(row[0])
        dates_month.append(row[1])
        dates_day.append(row[2])
        airT.append(row[3])
 
print(dates_year)
print(dates_month)
print(dates_day)
print(airT)
```
